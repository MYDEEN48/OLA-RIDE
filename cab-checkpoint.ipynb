```python


# Check the column names to confirm the ride status column
print(df.columns)

# Assuming the column is named 'ride_status'
status_counts = df['Booking _Status'].value_counts()

# Display the distribution
print(status_counts)

# Optional: Visualize it
plt.figure(figsize=(8, 5))
status_counts.plot(kind='bar', color='skyblue')
plt.title("Ride Status Distribution")
plt.xlabel("Status")
plt.ylabel("Number of Rides")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Filter only completed rides
df_completed = df[df["Booking _Status"] == "Completed"]

# Group by Payment Method and sum the Booking Value
revenue_by_payment = df_completed.groupby("Payment_Method")["Booking_Value"].sum()

# Display the result
print(revenue_by_payment)


plt.figure(figsize=(8, 5))
revenue_by_payment.plot(kind='bar', color='teal')
plt.title("Revenue by Payment Type")
plt.xlabel("Payment Method")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Filter only completed rides
df_completed = df[df["Booking _Status"] == "Completed"]

# Count pickup locations
top_pickups = df_completed["Pickup_Location"].value_counts().head(10)

# Display the result
print(top_pickups)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
top_pickups.plot(kind='barh', color='coral')
plt.title("Top 10 Pickup Locations")
plt.xlabel("Number of Rides")
plt.ylabel("Pickup Location")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

```

```python
import pandas as pd

# Convert 'Date' column to datetime format
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors='coerce')

# Filter only completed rides
df_completed = df[df["Booking _Status"] == "Completed"]

# Create a 'Month' column
df_completed["Month"] = df_completed["Date"].dt.to_period("M")

# Group by Month and sum Booking_Value
monthly_revenue = df_completed.groupby("Month")["Booking_Value"].sum()

# Display the result
print(monthly_revenue)



monthly_revenue.plot(kind='line', marker='o', figsize=(10, 6), color='darkorange')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python


# Convert 'Date' column to datetime format (optional but useful)
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors='coerce')

# Filter only completed rides
df_completed = df[df["Booking _Status"] == "Completed"]

# Select relevant numerical columns
cols = ["Avg_VTAT", "Avg_CTAT", "Booking_Value", "Ride_Distance", "Driver_Ratings", "Customer_Rating"]
df_numeric = df_completed[cols]

# Compute correlation matrix
corr_matrix = df_numeric.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap - Completed OLA Rides")
plt.tight_layout()
plt.show()
```
